```python
print('Seja bem vindo a Loja do Rodrigo de Lima Gonçalves !!') # IDENTIFICADOR PESSOAL

# INSERÇÃO DE VALORES 
valor_unitario = float(input('Insira o valor unitário do produto: ')) #VALOR DO PRODUTO 
qtd_produto = int(input('Insira a quantidade de produtos desejada: '))#VALOR DA QUANTIDADE DE PRODUTOS

# ESTRUTURA DE DECISÃO DO DESCONTO
if (qtd_produto <= 9): # 0 A 9 UNIDADES 0% DE DESCONTO
  desconto = 0
elif (10 <= qtd_produto <= 99): # 10 A 99 UNIDADES 5% DE DESCONTO
  desconto = 0.05
elif (100 <= qtd_produto <= 999): # 100 A 999 UNIDADES 10% DE DESCONTO
  desconto = 0.1
else:                               # ACIMA DE 1000 UNIDADES 15% DE DESCONTO
  desconto = 0.15

# CALCULO DOS VALORES FINAIS
valor_total = valor_unitario * qtd_produto # VALOR TOTAL SEM DESCONTO
valor_total_desconto = valor_total * (1 - desconto) # VALOR TOTAL COM O RESPECTIVO DESCONTO
desconto_porcento = desconto *100 # TRANSFORMAÇÃO DO DESCONTO DE DECIMAL PARA PORCENTAGEM

# SAIDA COM MENSAGEM E VALORES FINAIS A SEREM PAGOS
print('O valor total a ser pago sem desconto é de R$ {}'.format(valor_total)) #VALOR TOTAL SEM DESCONTO
print('O valor total a ser pago com desconto é de R$ {:.2f} (desconto {:.0f}%)'.format(valor_total_desconto, desconto_porcento)) # VALOR TOTAL COM DESCONTO E DESCONTO EM PORCENTAGEM
```

```python
print('Seja bem vindo a Lanchonete do Rodrigo de Lima Gonçalves !!') # IDENTIFICADOR PESSOAL
print('') 
print('================ CARDÁPIO =================') # Cardápio indicando o codigo, descrição e preço de cada produto
print('')
print(' Codigo |   Descrição             | Valor |')
print('  100   |   Cachorro-Quente       | 9,00  |')
print('  101   |   Cachorro-Quente Duplo | 11,00 |')
print('  102   |   X-Egg                 | 12,00 |')
print('  103   |   X-Salada              | 13,00 |')
print('  104   |   X-Bacon               | 14,00 |')
print('  105   |   X-Tudo                | 17,00 |')
print('  200   |   Refrigerante Lata     | 5,00  |')
print('  201   |   Chá Gelado            | 4,00  |')
print('===========================================')

# Defino os valores inicias para as variaveis
valor = 0
pedir_mais = ''

while pedir_mais != 'N': # Primeiro loop. Enquanto a variavel pedir_mais for diferente de 'N' o loop se mantém.
 
  codigo = input('Insira o código do produto desejado: ')

  # Verifica se o codigo inserido pelo usúario é valido, ou seja, se correspondo a algum produto do cardapio

  if codigo != '100' and codigo != '101' and codigo != '102' and codigo != '103' and codigo != '104' and codigo != '105' and codigo != '200' and codigo != '201':
    print('Você deve inserir um código de produto válido!')
    continue  # Caso o usúario insira um código de produto inválido o algoritmo reinicia o loop
  
  if codigo == '100': # Verifica espeficamente qual codigo foi selecionando e soma o valor do produto ao carrinho, representado pela variavel 'valor'
    valor += 9
    print('Você adicionou um Cachorro-Quente no valor de R$ 9,00 no seu carrinho!') 

  elif codigo == '101':
    valor += 11
    print('Você adicionou um Cachorro-Quente Duplo no valor de R$ 11,00 no seu carrinho!') 
  
  elif codigo == '102':
    valor += 12
    print('Você adicionou um X-Egg no valor de R$ 12,00 no seu carrinho!')  

  elif codigo == '103':
    valor += 13
    print('Você adicionou um X-Salada no valor de R$ 13,00 no seu carrinho!')  

  elif codigo == '104':
    valor += 14
    print('Você adicionou um X-Bacon no valor de R$ 14,00 no seu carrinho!')  

  elif codigo == '105':
    valor += 17
    print('Você adicionou um X-Tudo no valor de R$ 17,00 no seu carrinho!')  

  elif codigo == '200':
    valor += 5
    print('Você adicionou um Refrigerante Lata no valor de R$ 5,00 no seu carrinho!')  
  
  elif codigo == '201':
    valor += 4
    print('Você adicionou um Chá Gelado no valor de R$ 4,00 no seu carrinho!') 
  
  while True: # Segundo loop, Verifica se a resposta do Usúario é valida (S/N). 
    pedir_mais = input('Deseja adicionar outro produto ao carrinho (S/N)? ')
    pedir_mais = pedir_mais.upper()

    if pedir_mais == 'S': # se a reposta for Sim o programa reinicia o loop e pede para que o usuario insira um novo produto ao carrinho
        break
    elif pedir_mais == 'N': # se a reposta for Não o programa encerra o loop e indica o valor total a ser pago pelo usuario
        print('Seu pedido foi concluído com sucesso!!!')
        print('O valor total a ser pago é de R$ {:.2f}'.format(valor))
        break  
    else:
        print('Você deve digitar S para Sim ou N para Não.')
        continue  # caso a entrada seja inválida, o loop interno se reinicia   
  
```

```python
def volume_objeto():
    while True:
        try:
            largura = float(input('Insira a largura do objeto em centimetros: '))
            altura = float(input('Insira a altura do objeto em centimetros: '))
            comprimento = float(input('Insira a comprimento do objeto em centimetros: '))
            volume = largura * altura * comprimento
            if volume < 1000:
                return 10
            elif 1000 <= volume < 10000:
                return 20
            elif 10000 <= volume < 30000:
                return 30
            elif 30000 <= volume < 100000:
                return 50
            else:
                print('O volume do objeto ultrapassa o limite máximo permitido.')
                print('Desculpe o transtorno. Insira novamente as dimensões do objeto.')
        except ValueError:
            print("Por favor, digite um valor numérico válido para as dimensões do objeto.")


def peso_objeto():
    while True:
        try:
            peso = float(input("Digite o peso do objeto em kilogramas: "))
            if peso <= 0:
                print("O peso do objeto deve ser maior que zero.")
            elif peso < 0.1:
                return 1
            elif 0.1 <= peso < 1:
                return 1.5
            elif 1 <= peso < 10:
                return 2
            elif 10 <= peso < 30:
                return 3
            else:
                print("O peso do objeto ultrapassou o limite máximo de peso permitido.")
                print('Entre com o peso do objeto novamente')
        except ValueError:
            print("Por favor, digite um valor numérico válido para o peso do objeto.")


def rota_objeto():
    while True:
        origem = input('Insira a sigla do estado de origem do objeto: ').upper()
        destino = input('Insira a sigla do estado de destino do objeto: ').upper()
        rota = destino + '-' + origem
        if rota == ('RJ-SP' or 'SP-RJ'):
            return 1
        elif rota == ('DF-SP' or 'SP-DF'):
           return 1.2
        elif rota == ('DF-RJ' or 'RJ-DF'):
           return 1.5
        else:
           print('Por favor, digite uma rota válida.')


# Programa principal
volume = volume_objeto()
peso = peso_objeto()
rota = rota_objeto()

valor_total = volume * peso * rota
print('O valor total do objeto é R$ {:.2f}'.format(valor_total))
```
